# Analiza Wyników 

Cały kod do analizy językowej znajduje się w folderze "lang_analysis". Korzysta on z internetu aby pobrać artykuły - informacje co do wykorzystania kodu do analizy innych artykułów znajdują się w pliku language_analysis.py

Do przeprowadzenia analizy wybrano wartości K = [3, 10, 100, 1000] oraz języki: angielski, niemiecki i francuski. Testowano artykuły z Pokewiki w języku francuskim i niemieckim oraz z explainxkcd w języku angielskim. Wybrano też kilka dzieł literatury aby porównać wyniki. 

Metoda używana do porównania zgodności to KL Divergence: https://en.wikipedia.org/wiki/Kullback%E2%80%93Leibler_divergence

Foldery:


json/art zawiera teksty literatury


json/en artykuły z explainxkcd po angielsku


json/de artykuły z Pokewiki po niemiecku


json/fr artykuły z Pokewiki po francusku

Jak widać na wykresach - naturą metody KL divergence jest zmnięjszająca się zgodność wraz z wydłużaniem się ogona mało prawdopodobnych zdażeń. Jednocześnie odpowiednie języki wykazują ciągle o wiele lepszą zgodność niż nieodpowiednie. Zauważmy też, że język niemiecki naturalnie przejął wiele słów z angielskiego i zatem wykazuje średnio o wiele lepsza zgodność z json/en niż francuski - choć ogólnie język angielski, ponieważ jest miksem francuskiego i niemieckiego, wykazuje dobrą zgodność z obydwoma swoimi kuzynami.

# Wnioski

Podczas przeprowadzania analizy napotkałem się na kilka metod zaburzających wyniki:

- Stosowanie least squares, szczególnie na skali logarytmicznej, dające totalnie bezsensowny rezultat
- Niezadbanie o format utf-8 sprawiające, że angielski był o wiele bardziej podobny niż inne języki nawet do Pana Tadeusza, ponieważ angielski to tak naprawdę zupa z ASCII
- Błędy w normalizacji, które faworyzowały małą liczbę danych - to znaczy dawały bonus za słowa, których nie było w tekście

Wszelkie te błędy były nieoczywiste i taka analiza okazała się trudniejsza do przeprowadzania niż myślałem.
